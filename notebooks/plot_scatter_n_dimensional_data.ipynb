```python
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets.samples_generator import make_blobs
```

```python
x, y = make_blobs(n_samples=200, centers=3, n_features=2, random_state=0)

plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

## Plot n-dimensional data with Principal Component Analysis (PCA)

```python
x_norm = (x - x.min())/(x.max() - x.min())

pca = sklearnPCA(n_components=2) #2-dimensional PCA
transformed = pca.fit_transform(x_norm)


plt.scatter(transformed[y==0][:,0], transformed[y==0][:,1], label='Class 1', c='red')
plt.scatter(transformed[y==1][:,0], transformed[y==1][:,1], label='Class 2', c='blue')
plt.scatter(transformed[y==2][:,0], transformed[y==2][:,1], label='Class 3', c='lightgreen')

plt.legend()
plt.show()
```
